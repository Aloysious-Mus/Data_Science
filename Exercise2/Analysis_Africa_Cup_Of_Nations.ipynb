```python
# Import Required Libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline


# Load Dataset
df = pd.read_csv('AfricaCupofNations.csv')
df.head()
```

```python
df.head(7)#the first 7 rows
```

```python

# Select the specified columns
selected_columns = df[['HomeTeam', 'AwayTeam', 'HomeTeamGoals', 'AwayTeamGoals']]

# Display the result
print(selected_columns)
```

```python
# Select rows where Egypt is either HomeTeam or AwayTeam
egypt_matches = df[(df['HomeTeam'] == 'Egypt') | (df['AwayTeam'] == 'Egypt')]

# Display the results
print("Matches involving Egypt:")
print(egypt_matches)
```

```python
# Number of rows
num_rows = len(df)

# Number of columns
num_columns = len(df.columns)

print(f"Number of rows: {num_rows}")
print(f"Number of columns: {num_columns}")
```

```python
# Select rows where 'Attendance' is missing
missing_attendance = df[df['Attendance'].isna()]  # .isna() and .isnull() are equivalent

# Display the result
print(f"Rows with missing Attendance: ({len(missing_attendance)} found)")
print(missing_attendance)
```

```python
 #Select rows where 'HomeTeamGoals' is between 3 and 6 inclusive
result = df[df['HomeTeamGoals'].between(3, 6)]

# Display the result
print(result)
```

```python
# Change "AwayTeamGoals" in the 3rd row (position index 2) to 10
# Using position-based indexing (.iloc)
df.iloc[2, df.columns.get_loc('AwayTeamGoals')] = 10

# Display modified DataFrame
print("\nData after modification:")
print(df)
```

```python
# Sort the DataFrame
sorted_df = df.sort_values(
    by=['HomeTeam', 'HomeTeamGoals'],
    ascending=[True, False]  # HomeTeam ascending, HomeTeamScores descending
)

# Display sorted DataFrame
print("Sorted DataFrame:")
print(sorted_df)
```

```python
# Get list of column headers
column_headers = df.columns.tolist()

# Display the result
print("Column Headers:", column_headers)
```

```python
#add column for total goals
#and Also add two rows 
if 'HomeTeamGoals' in df and 'AwayTeamGoals' in df:
    df['TotalGoals'] = df['HomeTeamGoals'] + df['AwayTeamGoals']
    
    # Display modified DataFrame
print("\nDataFrame with new column:")
print(df.head())

# Display new column list
print("\nUpdated columns:", df.columns.tolist())
```

```python
# Create two empty rows as a new DataFrame
empty_rows = pd.DataFrame(
    np.nan,  # Fill with NaN values
    index=range(2),  # Two rows
    columns=df.columns  # Same columns as original
)

# Append empty rows to original DataFrame
df_with_empty = pd.concat([df, empty_rows], ignore_index=True)

# Display modified DataFrame
print("\nDataFrame with 2 empty rows added:")
print(df_with_empty.tail(3))  # Show last 3 rows (includes new empty rows)

```

```python

# Replace 'Uganda' with 'China' in 'AwayTeam' column
df['AwayTeam'] = df['AwayTeam'].replace('Uganda', 'China')

# Display modified values
print("\nModified 'AwayTeam' values:")
print(df['AwayTeam'].value_counts())
```
